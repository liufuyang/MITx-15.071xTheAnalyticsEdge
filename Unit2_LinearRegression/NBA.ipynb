## Build Regression - PointReg

```r
PointsReg = lm(PTS ~ X2PA + X3PA + FTA + AST + ORB + DRB + TOV + STL + BLK, data=NBA)
```

```r
summary(PointsReg)
```

```r
PointsReg$residuals
```

```r
SSE = sum(PointsReg$residuals^2)
```

```r
SSE
```

```r
RMSE = sqrt(SSE/nrow(NBA))
```

```r
RMSE
```
