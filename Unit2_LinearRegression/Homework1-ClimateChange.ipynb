```r
climate_change = read.csv('data/climate_change.csv')
```

```r
str(climate_change)
```

```r
summary(climate_change)
```

```r
climate_train_data = subset(climate_change, Year <= 2006)
climate_test_data = subset(climate_change, Year > 2006)
```

```r
?lm
```

```r
model1 = lm(Temp ~ MEI + CO2 + CH4 + N2O + CFC.11 + CFC.12 + TSI + Aerosols, data=climate_train_data)
```

```r
summary(model1)
```

```r
cor(climate_train_data)
```

## N20 only model:

```r
model_N20 = lm(Temp ~ MEI + N2O + TSI + Aerosols, data=climate_train_data)
summary(model_N20)
```

## Try the Step function:

The step function has one argument - the name of the initial model. It returns a simplified model. Use the step function in R to derive a new model, with the full model as the initial model (HINT: If your initial full model was called "climateLM", you could create a new model with the step function by typing step(climateLM). Be sure to save your new model to a variable name so that you can look at the summary. For more information about the step function, type ?step in your R console.)

```r
?step
```

```r
model_s = step(model1)
```

```r
summary(model1)
```

```r
summary(model_s)
```

## Testing on Unseen Data:

```r
predictions = predict(model_s, newdata = climate_test_data)
```

```r
SSE = sum((predictions - climate_test_data$Temp)^2)
SST = sum((mean(climate_train_data$Temp) - climate_test_data$Temp)^2)

1 - SSE/SST
```
